```python
import numpy as np
import pandas as pd
import nltk
nltk.download('stopwords')
import gensim
import multiprocessing
print multiprocessing.cpu_count()
```

```python
train = pd.read_csv('gnm_articles.csv')
```

```python
train['article_text'] = train['article_text'].str.replace('<p>','')
train['article_text'] = train['article_text'].str.replace('[^a-zA-Z]',' ').str.lower()
```

```python
train['article_text'].head(10)
```

```python
stop_re = '\\b'+'\\b|\\b'.join(nltk.corpus.stopwords.words('english'))+'\\b'
train['article_text'] = train['article_text'].str.replace(stop_re, '')
```

```python
train['article_text'].head(10)
```

```python
train['article_text'] = train['article_text'].str.split()
```

```python
train['article_text'].head(10)
```

```python
phrases = gensim.models.phrases.Phrases(train['article_text'].tolist())
phraser = gensim.models.phrases.Phraser(phrases)
train_phrased = phraser[train['article_text'].tolist()]
```

```python
w2v = gensim.models.word2vec.Word2Vec(sentences=train_phrased,workers=4)
```

```python
w2v.wv.save('SFUvecs')
```

```python
w2v.most_similar('climate')
```
