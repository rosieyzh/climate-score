```python
import numpy as np
import pandas as pd
import gensim
import nltk
nltk.download('stopwords')
```

```python
file = pd.read_csv('data.csv')
file.head(10)
```

```python
articles = file['content']

'''
Pre-processing:
>only retain words
>convert to lowercase
>remove stopwords
'''
articles = articles.str.replace('[^a-zA-Z]',' ').str.lower()
stop_re = '\\b'+'\\b|\\b'.join(nltk.corpus.stopwords.words('english'))+'\\b'
articles = articles.str.replace(stop_re, '')
articles = articles.str.split()
# print(articles)
```

```python
cleaned_articles = [x for x in articles if str(x) != 'nan']
```

```python
#Input articles through Phraser
phrases = gensim.models.phrases.Phrases(cleaned_articles)
phraser = gensim.models.phrases.Phraser(phrases)
train_phrased = phraser[cleaned_articles]
```

```python
#Methods to calculate tf-idf scores
def word_count(text):
    '''
    Input:
    text - list object of tokenized strings
    '''
    return len(text)

def freq_count(articles):
    '''
    Input:
    articles - list of list of tokenized words
    Output:
    freq_dict_list - list of dictionary containing frequencies of each word in text
    '''
    freq_dict_list = []
    doc_id=0
    for text in articles:
        doc_id+=1
        freq_dict = {}
        for word in text:
            if word in freq_dict:
                freq_dict[word] += 1
            else:
                freq_dict[word] = 1
        doc_temp = {'id' : doc_id, 'freq_dict' : freq_dict }
        freq_dict_list.append(doc_temp)
    
    return freq_dict_list
    
```

```python
def computeDocTFDict(doc):
    """ Returns a tf dictionary for each doc whose keys are all 
    the unique words in the doc and whose values are their 
    corresponding tf.
    """
    #Counts the number of times the word appears in doc
    docTFDict = {}
    for word in doc:
        if word in docTFDict:
            docTFDict[word] += 1
        else:
            docTFDict[word] = 1
    #Computes tf for each word           
    for word in docTFDict:
        docTFDict[word] = docTFDict[word] / len(doc)
    return docTFDict
```

```python
# result_docTFDict = computeDocTFDict(cleaned_articles[0])
result_docTFDict = []
for i in range(len(cleaned_articles)):
    result_docTFDict.append(computeDocTFDict(cleaned_articles[i]))
```

```python
print(result_docTFDict[0])
```

```python
def computeCountDict():
    """ Returns a dictionary whose keys are all the unique words in
    the dataset and whose values count the number of reviews in which
    the word appears.
    """
    countDict = {}
    # Run through each review's tf dictionary and increment countDict's (word, doc) pair
    for doc in result_docTFDict:
        for word in doc:
            if word in countDict:
                countDict[word] += 1
            else:
                countDict[word] = 1
    return countDict

#Stores the review count dictionary
countDict = computeCountDict()
```

```python
countDict['winter']
```

```python
import math

def computeIDFDict():
    """ Returns a dictionary whose keys are all the unique words in the
    dataset and whose values are their corresponding idf.
    """
    idfDict = {}
    for word in countDict:
        idfDict[word] = math.log(len(cleaned_articles) / countDict[word])
    return idfDict

#Stores the idf dictionary
idfDict = computeIDFDict()
```

```python
idfDict['winter']
```

```python
def computeDocTFIDFDict(docTFDict):
    """ Returns a dictionary whose keys are all the unique words in the
    doc and whose values are their corresponding tfidf.
    """
    docTFIDFDict = {}
    #For each word in the review, we multiply its tf and its idf.
    for word in docTFDict:
        docTFIDFDict[word] = docTFDict[word] * idfDict[word]
    return docTFIDFDict

#Stores the TF-IDF dictionaries
tfidfDict = [computeDocTFIDFDict(doc) for doc in result_docTFDict]
```

```python
extreme_weather_seed = ['natural_disaster', 'extreme_weather', 'wildfire', 'heat_wave', 'drought',
                'forest_fire', 'flood', 'flash_flood', 'hurricane', 'earthquake', 'tsunami', 'cyclone']
```

```python
def calcXWeatherTFIDF():
    
    totalXWeatherScore = [0]*(len(tfidfDict))
                
    for i in range(len(tfidfDict)):
        temp_dict = tfidfDict[i]
        for word in temp_dict.keys():
            if word in extreme_weather_seed:
                print(word)
                totalXWeatherScore[i] += temp_dict[word]
                
    return totalXWeatherScore
```
