```python
# from nltk.parse import CoreNLPParser
from pycorenlp import StanfordCoreNLP

nlp = StanfordCoreNLP('http://localhost:9000')
```

```python
import numpy as np
import pandas as pd

data = pd.read_csv('news-data.csv', error_bad_lines=False);
data.head(5)
```

```python
def get_sentiments(article):
    result = nlp.annotate(article,
                   properties={
                       'annotators': 'sentiment',
                       'outputFormat': 'json',
                       'timeout': 80000,
                   })
    return result

def print_sentiment(title, result):
    average_score = 0
    num_sentences = len(result["sentences"])
    sentiments = ["Very negative", "Negative", "Neutral", "Positive", "Very positive"]

    for s in result["sentences"]:
        average_score += int(s["sentimentValue"])
    
    average_score /= num_sentences
    print('{}: {}, {}'.format(title, average_score, sentiments[round(average_score)]))
    
def print_sentence_sentiment(title, result):
    average_score = 0
    num_sentences = len(result["sentences"])
    sentiments = ["Very negative", "Negative", "Neutral", "Positive", "Very positive"]
    
    for s in result["sentences"]:
        print(' '.join([t["word"] for t in s["tokens"]]))
        print('Value:{}, Sentiment:{}'.format(s["sentimentValue"], s["sentiment"]))
        print()
        average_score += int(s["sentimentValue"])
        
    average_score /= num_sentences
    print('{}: {}, {}'.format(title, average_score, sentiments[round(average_score)]))
```

```python
#Goes through the given articles and calculates their average sentiments
for i in range(50, 55):
    article = data["content"][i]
    title = data["title"][i]
    result = get_sentiments(article)
    print_sentiment(title, result)
    
```
